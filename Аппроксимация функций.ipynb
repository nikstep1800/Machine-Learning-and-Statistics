```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import interpolate
```

# Коэффициенты 1, 15


## Многочлен 1 степени

```python
x = np.array([1, 15])
y = np.sin(x/5.)*np.exp(x/10.)+5*np.exp(-x/2.)
y1 = y[0]
y2 = y[1]
a1 = np.array([[1, 1],[1, 15]])
b1 = np.array([y1, y2])
w1 = np.linalg.solve(a1, b1)
print(w1)
```

## График

```python
x = np.arange(0, 16, 0.5)
y = np.sin(x/5.)*np.exp(x/10.)+5*np.exp(-x/2.)
q = w1[0] + w1[1]*x
plt.plot(x, y)
plt.plot(x, q)
```

# Коэффициенты 1, 8, 15


## Многочлен 2 степени

```python
x = np.array([1, 8, 15])
y = np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2)
y1 = y[0]
y2 = y[1]
y3 = y[2]
a2 = np.array([[1, 1, 1],[1, 8, 8**2], [1, 15, 15**2]])
b2 = np.array([y1, y2, y3])
w2 = np.linalg.solve(a2, b2)
print(w2)
```

## График

```python
x = np.arange(0, 16, 0.5)
y = np.sin(x/5.)*np.exp(x/10.)+5*np.exp(-x/2.)
q = w2[0] + w2[1]*x + w2[2]*x**2
plt.plot(x, y)
plt.plot(x, q)
```
